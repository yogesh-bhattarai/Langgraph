```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"/content/summary_2024-03-31.csv")
```

```python
df
```

```python
top_symbol= df.groupby("symbol")['total_buy_qty'].sum().sort_values(ascending=False).head(10)
```

```python
plt.xticks(rotation=45)
top_symbol.plot(kind='bar', figsize=(10,6), color='skyblue')
```
